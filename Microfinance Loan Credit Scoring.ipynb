```python
df = df.replace(',', '', regex=True)
df=df.replace('-', np.nan, regex=True)
df = df.apply(pd.to_numeric).astype(float)
```

```python
def calculate_risk_score(row):
    # 1. Consistency: Percentage of weeks with payments
    total_weeks = len(row) - 1  # Exclude 'Farmer' column
    payment_weeks = row[1:].count()  # Count non-NaN values
    consistency = (payment_weeks / total_weeks) * 100
    consistency_score = 100 - consistency  # Lower consistency = higher risk

    # 2. Variability: Standard deviation of payments
    payments = row[1:].dropna()
    if len(payments) > 0:
        variability = np.std(payments) / np.mean(payments)
        variability_score = variability * 20  # Scale for scoring
    else:
        variability_score = 50  # High penalty for no data

    # Total risk score
    risk_score = consistency_score + variability_score
    return risk_score
```

```python
df['Risk_Score'] = df.apply(calculate_risk_score, axis=1)
```

```python
df['Risk_Score'].max()
```

```python
import seaborn as sns 

plt.hist(df['Risk_Score'], bins=50)
plt.title('Distribution of Risk Scores Across Farmers')
plt.xlabel('Risk Score')
plt.ylabel('Number of Farmers')
plt.grid(True, alpha=0.8)
plt.xticks(np.arange(0, 110, 3))
plt.show()
```

```python
df['Risk_Level'] = pd.cut(df['Risk_Score'], bins=[0, 31, 56, float('inf')], labels=['Low', 'Moderate', 'High'])
```
